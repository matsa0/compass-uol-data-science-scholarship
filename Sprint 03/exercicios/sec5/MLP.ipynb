```python
from sklearn import datasets
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target,
                                                                  test_size=0.3, random_state=0)
```

### Definindo hiperparâmetros

```python
model = MLPClassifier(
    verbose=True, # acompanhamento iterativo
    hidden_layer_sizes=(5,4), # 2 camadas ocultas, uma com tamanho 5 e outra 4
    activation='relu', # o que faz a rede neural funcionar 
    batch_size=20,
    learning_rate='adaptive',
    momentum=0.9,
    early_stopping=False,
    max_iter=1000, #quantidade de vezes(epochs) que os dados serão propagados
    random_state=10
)

model.fit(X_train, y_train)
```

### Visualizado a loss function 

```python
plt.plot(model.loss_curve_)
plt.xlabel('Iterações')
plt.ylabel('Valor de loss')
plt.title('Loss Rate do modelo MLP')
```

```python
previsions = model.predict(X_test)
previsions
```

```python
print(accuracy_score(y_test, previsions))
```
