```python
import numpy as np
import matplotlib.pyplot as plt
```

# Learning about pixels

```python
arr = np.arange(0, 256, 1)
arr
```

```python
arr = arr.reshape((16, 16))
arr
```

```python
arr_8bit = arr.astype(np.uint8)
plt.imshow(arr_8bit, cmap='gray')
plt.title("256 pixels em matrizes 16x16")
plt.axis('off')
plt.show()
```

```python
arr1 = np.random.randint(low=0,high=255,size=(16,16),dtype=np.uint8) #matriz 16x16 aleatória
plt.imshow(arr1,cmap='gray')
plt.axis('off')
plt.show()
```

# Reading image

```python
import cv2
import matplotlib.image as matimg
from PIL import Image

import matplotlib.pyplot as plt
```

```python
# Diferentes formas de ler uma image

img_cv = cv2.imread('../data/flying_bird.jpg') # retorna NumPy (BGR) - OpenCV

img_mat = matimg.imread('../data/flying_bird.jpg') # Exibição de imagens com Matplotlib

img_pil = Image.open('../data/flying_bird.jpg') # Objeto PIL (RGB) - Manipulação e edição de imagens
```

```python
print(img_cv)
# B G R
```

```python
print(img_mat)
# R G B
```

```python
img_pil
```

# Display Image

```python
# Using opencv
plt.imshow(img_cv)
plt.axis('off')
plt.title('OpenCV BGR')
plt.show()
```

```python
# Using Matplotlib
plt.imshow(img_mat)
plt.axis('off')
plt.title('Matplotlib RGB')
plt.show()
```

```python
# Using Pillow
plt.imshow(img_pil)
plt.axis('off')
plt.title('Pillow RGB')
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.subplot(1,3,1)
plt.imshow(img_cv)
plt.axis('off')
plt.title('OpenCV BGR')

plt.subplot(1,3,2)
plt.imshow(img_mat)
plt.axis('off')
plt.title('Matplotlib RGB')

plt.subplot(1,3,3)
plt.imshow(img_pil)
plt.axis('off')
plt.title('Pillow RGB')

plt.show()
```

# Color Space

Color conversion<br>
Gray scale

```python
img_cv.shape #rows, colors, channels
```

```python
# b g r
b = img_cv[:,:,0]
g = img_cv[:,:,1]
r = img_cv[:,:,2]
```

```python
plt.figure(figsize=(15,8))
plt.subplot(1,4,1)
plt.imshow(b)
plt.axis('off')
plt.title('Blue Channel')

plt.subplot(1,4,2)
plt.imshow(g)
plt.axis('off')
plt.title('Green Channel')

plt.subplot(1,4,3)
plt.imshow(r)
plt.axis('off')
plt.title('Red Channel')

plt.subplot(1,4,4)
plt.imshow(img_mat)
plt.axis('off')
plt.title('Image RGB')

plt.show()
```

```python
#Conversão de BGR para RGB
img_cv_rgb = cv2.cvtColor(img_cv,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_cv_rgb)
plt.axis('off')
plt.show()
```
