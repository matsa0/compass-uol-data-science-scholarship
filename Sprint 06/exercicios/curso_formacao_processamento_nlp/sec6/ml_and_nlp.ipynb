# Classificação de Spam

```python
spam = pd.read_csv("spam.csv")
spam.head()
```

```python
spam['Category'].value_counts()
```

A classe está desbalanceada, existem muito mais categorias ham(legítima) do que spam

```python
spam.dtypes
```

```python
spam.shape
```

```python
X = spam['Message'] #independente
y = spam['Category'] #dependente
```

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(X)
X.shape
```

O **TfidfVectorizer** converte texto em uma matriz numérica, tornando os dados adequados para modelos de Machine Learning.

Ele transforma texto em números usando a técnica TF-IDF (Term Frequency - Inverse Document Frequency). O objetivo é **dar mais peso a palavras importantes e reduzir a influência de palavras comuns** (stopwords).

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_test.shape
```

```python
X_train.shape
```

```python
y_test.shape
```

```python
y_train.shape
```

```python
rf = RandomForestClassifier(n_estimators=500, verbose=2, random_state=42)
rf.fit(X_train, y_train)
```

```python
prev = rf.predict(X_test)
print(prev)
```

```python
print(confusion_matrix(y_test, prev))
```

- 1448 Verdadeiros Negativos (TN): O modelo previu corretamente que 1448 mensagens eram ham.
- 186 Verdadeiros Positivos (TP): O modelo previu corretamente que 186 mensagens eram spam.
- 38 Falsos Negativos (FN): O modelo previu ham, mas na verdade eram spam (ou seja, deixou passar spams).
- 0 Falsos Positivos (FP): O modelo não classificou nenhuma mensagem ham como spam.

```python
print(metrics.classification_report(y_test, prev))
```
