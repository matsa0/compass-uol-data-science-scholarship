# Análise de Sentimento com o Vader

O ** no VADER (Valence Aware Dictionary and sEntiment Reasoner) é uma métrica que representa o sentimento geral de um texto. É um valor real entre -1 (muitonegativo) e 1 (muito positivo). Calculada através das palavras da sentença


```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
mas = SentimentIntensityAnalyzer()
```

```python
a = 'I HATE this movie!!!'
x = mas.polarity_scores(a)
print(x)
```

```python
a = 'I hate this movie, but i liked the soundtrack.'
x = mas.polarity_scores(a)
print(x)
```

```python
a = 'This ice cream made my day happier'
x = mas.polarity_scores(a)
print(x)
```

```python
print(type(x))
```

```python
print(x['pos'])
```

**Polarity**: Mede o grau de positividade ou negatividade de um texto. Valor real entre -1 (muito negativo) e 1 (muito positivo)

**Subjectivity**: O quanto a sentença se refere a **opinião**, **julgamento** ou **emoção**. Valor real entre 0 e 1, quando mais próximo de 1, mais subjetivo
