# Tokens e StopWords

```python
text = "Nós somos feitos de poeira de estrelas. Nós somos uma maneira de o cosmos se autoconhecer. A imaginação nos leva a mundos que nunca sequer existiram. Mas sem ela não vamos a lugar algum."
print(text)
```

```python
sentences = sent_tokenize(text, language="portuguese")
print(type(sentences))
print(len(sentences))
print(sentences)
```

```python
tokens = word_tokenize(text, language="portuguese")
print(tokens)
print(len(tokens))
```

```python
stops = stopwords.words("portuguese")
print(len(stops))
print(stops)
```

```python
relevant_words = []

for token in tokens:
  if token not in stops:
    relevant_words.append(token)
```

```python
relevant_words, len(relevant_words)
```

```python
print(string.punctuation)
```

```python
relevant_words = [token for token in relevant_words if token not in string.punctuation]
relevant_words
```

```python
len(relevant_words)
```

# Produzindo métricas

```python
frequency = nltk.FreqDist(relevant_words)
frequency
```

```python
most_common = frequency.most_common(5)
most_common
```

# Stemming

Processo de analisar cada palavra  reduzi-la à sua raiz(seu significado). Uma característica dessa técnica é que ela pode reduzir a palavra a uma outra gramaticalmente incorreta, porém ainda com valor para nossa análise.

- Porter: Mais comum e mais antigo
- Snowball: Melhor performance computacional
- Lancaster: Mais agressivo. Resultados as vezes não intuitivos

```python
p_stemmer = PorterStemmer()

# aplica o stem a cada palavra relevante
p_stem_list = [p_stemmer.stem(word) for word in relevant_words]

print("Palavras sem stemmer: \n", relevant_words)
print("\nPalavras com PorterStemmer: \n", p_stem_list)
```

```python
snow_stemmer = SnowballStemmer("portuguese")
snow_stem_list = [snow_stemmer.stem(word) for word in relevant_words]

print("Palavras sem stemmer: \n", relevant_words)
print("\nPalavras com SnowballStemmer: \n", snow_stem_list)
```

```python
lanc_stemmer = LancasterStemmer()
lanc_stem_list = [lanc_stemmer.stem(word) for word in relevant_words]

print("Palavras sem stemmer: \n", relevant_words)
print("\nPalavras com LancasterStemmer: \n", lanc_stem_list)
```

# Lemmatization

busca simplificar a análise textual ao normalizar palavras para sua forma base, conhecida como “lema”. O objetivo é normalizar termos flexionais ou derivados em sua forma canônica, facilitando operações posteriores de busca, indexação ou classificação.

```python
lemmatizer = WordNetLemmatizer()
```

```python
result = [lemmatizer.lemmatize(word) for word in relevant_words]
print(relevant_words)
print(result)
```
