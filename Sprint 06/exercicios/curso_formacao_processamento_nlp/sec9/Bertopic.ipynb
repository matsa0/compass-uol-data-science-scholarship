```python
import glob
data = []
dir = "/content/documentos/"
```

```python
for f in glob.iglob(dir + '*.txt', recursive=False):
  myfile = open(f, 'r', encoding='utf8')
  content = myfile.read()

  data.append(content)
  myfile.close()
```

```python
sentences = []
for item in data:
  for token in sent_tokenize(item):
    sentences.append(token)
```

# BERTopic

O BERTopic é um modelo avançado de modelagem de tópicos que combina embeddings de linguagem (como BERT) com algoritmos de clusterização para identificar tópicos de maneira mais precisa



```python
model = BERTopic(language="portuguese", verbose=True,
                 top_n_words=15, #quantas palavras terá cada tópico
                 min_topic_size=10, #tamanho mínimo que um tópico deve ter
                 nr_topics=20, #reduz o número de tópcios iniciais para este número
                 vectorizer_model=CountVectorizer(ngram_range=(1,3), #vetorização
                  stop_words=stopwords.words("portuguese"), min_df=10))
```

```python
%%time
topics, prob = model.fit_transform(sentences)
```

```python
frequency = model.get_topic_info()
frequency.head(10)
```
