### Configuraçãoes do SageMaker

```python
import sagemaker
import boto3
from sagemaker import Session
```

```python
session = sagemaker.Session()
```

```python
bucket = 'aws-sagemaker-course-633'
subdir = 'models/house-prices/linear-learner'
subdir_dataset = 'datasets/house-prices'
```

```python
role = sagemaker.get_execution_role()
print(role)
```

```python
### Preparação dos dados para serem enviados para o SageMaker

import io
import sagemaker.amazon.common as smac #sagemaker commom library

buffer = io.BytesIO() #formato de bytes
smac.write_numpy_to_dense_tensor(buffer, X_train, y_train)
buffer.seek(0) #coloca na posição inicial
```

```python
import os
key = 'houses-train-data'
boto3.resource('s3').Bucket(bucket).Object(os.path.join(subdir_dataset, 'train', key)).upload_fileobj(buffer)
```

```python
os.path.join(subdir_dataset, 'train', key)
```

```python
s3_train_data = 's3://{}/{}/train/{}'.format(bucket, subdir_dataset, key)
```

```python
output_location = 's3://{}/{}/output'.format(bucket, subdir)
```

### Criação e treinamento do Linear Learner

```python
container = sagemaker.image_uris.retrieve(framework='linear-learner', region=boto3.Session().region_name)
```

```python
boto3.Session().region_name
```

```python
linear = sagemaker.estimator.Estimator(image_uri=container, # container
                                      role=role, # IAM role
                                      instance_count=1, # n de instancias EC2 (1 servidor)
                                      instance_type='ml.m5.large',
                                      output_path=output_location,
                                      sagemaker_session=session,
                                      use_stop_instances = True) 
```

```python
linear.set_hyperparameters(feature_dim=16, #16 colunas na base de dados
                          predictor_type='regressor',
                          num_models=16) 
```

```python
linear.fit({'train': s3_train_data})
```

### Deploy, previsões e avaliação

criação de um modelo que pode ser usado em produção para realização de previsões

```python
linear_regressor = linear.deploy(initial_instance_count=1, instance_type='ml.m5.large')
```

```python
from sagemaker.serializers import CSVSerializer
from sagemaker.deserializers import JSONDeserializer
```

```python
linear_regressor.serializer = CSVSerializer() #transforma os dados Python (como listas, arrays NumPy, ou Pandas DataFrames) em strings no formato CSV.
linear_regressor.deserializer = JSONDeserializer() #converte o JSON retornado pelo modelo (formato padrão para respostas do SageMaker) em objetos Python, como dicionários ou listas.
```

```python
results = linear_regressor.predict(X_test)
```

```python
results
```

```python
previsions = np.array([r['score'] for r in results['predictions']])
previsions
```

```python
previsions.shape, X_test.shape
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mae = mean_absolute_error(y_test, previsions)
mse = mean_squared_error(y_test, previsions)
print('MAE = ', mae, '\nMSE = ', mse)
```
