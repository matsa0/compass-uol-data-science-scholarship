```python
import pandas as pd
import plotly.express as px
```

```python
df_credit = pd.read_csv('../data/credit_data.csv')
```

```python
df_credit
```

```python
df_credit.isnull().sum()
```

```python
df_credit.dropna(inplace=True)
```

# Boxplot

```python
# Outliers idade
g = px.box(df_credit, y='age', height=600, width=600)
g.show()
#outliers - idades negativas?
```

```python
outliers_age = df_credit[df_credit['age'] < 0]
outliers_age
```

```python
# Outliers loan
g = px.box(df_credit, y='loan', height=700, width=600)
g.show()
#clientes com dÃ­vida muito alta
```

```python
outliers_loan = df_credit[df_credit['loan'] > 13300]
outliers_loan
```

# Scatter

```python
# Income x age
g = px.scatter(x = df_credit['income'], y = df_credit['age'])
g.show()
```

```python
# Income x loan
g = px.scatter(x = df_credit['income'], y = df_credit['loan'])
g.show()
```
