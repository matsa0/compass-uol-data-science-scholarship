```python
import matplotlib.pyplot as plt
```

```python
cidade_a_temperaturas = [32,30,27,28,29,24]
cidade_b_temperaturas = [27,26,29,25,22,22]
cidade_c_temperaturas = [17,19,15,20,25,26]
datas = [5,10,15,20,25,30]
```

```python
#Colocar uma barra ao lado da outra
barras_a = list(range(len(datas))) #cria uma lista com o tamanho das datas
barras_b = [barra + 0.25 for barra in barras_a]#desloca a barra b para o lado da barra a
barras_c = [barra + 0.25 for barra in barras_b]#desloca a barra c para o lado da barra b

#eixo x, eixo y
graf_barra_cidade_a = plt.bar(barras_a, cidade_a_temperaturas, width=0.25, label='Cidade A')
graf_barra_cidade_b = plt.bar(barras_b, cidade_b_temperaturas, width=0.25, label='Cidade B')
graf_barra_cidade_c = plt.bar(barras_c, cidade_c_temperaturas, width=0.25, label='Cidade C')


plt.title('Temperatura máxima das cidades')
plt.xlabel('Dia do mês')
plt.ylabel('Temperatura')
plt.legend(title='cidades', loc='best', bbox_to_anchor=(1, 0.9))

#ticks='valor antigo', labels='valor novo'
x = plt.xticks(ticks=barras_a, labels=datas)

plt.savefig('grafico_barras.png')
```
