```python
from matplotlib import pyplot as plt
```

```python
eixo_x_dias = [1,5,10,15,20,25,30]

eixo_y_temp_max = [28,29,25,32,34,36,31]
eixo_y_temp_min = [21,22,17,23,23,24,20]
```

```python
plt.title('Temperaturas máximas e mínimas')
plt.xlabel('Dia do mês')
plt.ylabel('Temperatura')

#Temperatura máxima
plt.plot(eixo_x_dias, eixo_y_temp_max, linestyle='--', marker='o', color='orangered', label='Temp Máx') 
#Temperatura mínima
plt.plot(eixo_x_dias, eixo_y_temp_min, marker='o', color='deepskyblue', label='Temp mín') 
plt.legend(loc='upper left')
plt.grid(True)

#onde começa o eixo x, onde termina o eixo x, onde começa o eixo y, onde termina o eixo y
print(plt.axis())
plt.axis([1, 31, 5, 45])

plt.savefig('meu_primeiro_grafico.png')
```
