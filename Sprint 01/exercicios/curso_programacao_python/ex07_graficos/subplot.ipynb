```python
import matplotlib.pyplot as plt

#Figura: o espaço que pode ser preenchido com informações. Toda a janela na interface do usuário 
#Eixos: lugar onde os dados são representados, geralmente os eixos x e y.
#Subplot: uma figura com vários gráficos
```

```python
# cria uma figura e uma grade de subplots
fig, axs = plt.subplots(nrows=1, ncols=2)

#adicionando dados ao subplot esquerdo
axs[0].plot([10,15,30,20,8], [5,10,15,20,30], label='Esquerda')
axs[0].set_title('Subplot do lado esquedo')

#adicionando dados ao subplot direito
axs[1].barh(['A','B','C','D'], [8,15,6,7], color='orange')
axs[1].set_title('Subplot do lado direito')

#Título geral para a figura
fig.suptitle('Aula de Subplots com Matplotlib')
fig.tight_layout(pad=2.0) #ajusta automaticamente o layout(espaçamento) da figura
```

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10,8)) #figsize(polegadas)tamanho da moldura

fig.suptitle('Aula de Subplots com Matplotlib')
fig.tight_layout(pad=3.0) 

#Subplots
#adicionando dados ao subplot superior esquerdo
axs[0, 0].plot([10,15,30,20,8], [5,10,15,20,30], label='Esquerda') #primeira linha, primeira coluna
axs[0, 0].set_title('Subplot do lado superior esquerdo')

#adicionando dados ao subplot superior direito
axs[0,1].bar(['A','B','C','D'], [8,15,6,7], color='green')
axs[0,1].set_title('Subplot do lado superior direito')

#adicionando dados ao subplot inferior esquerdo
axs[1,0].barh(['A','B','C','D'], [8,15,6,7], color='orange')
axs[1,0].set_title('Subplot do lado inferior esquerdo')

#adicionando dados ao subplot inferior direito
axs[1,1].pie([30,20,40], labels=['A', 'B', 'C'])
axs[1,1].set_title('Subplot do lado inferior direito')

plt.savefig('subplots.png')
```
